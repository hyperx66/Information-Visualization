# Data Processing

Prepare CSV file for reading

```python
file = open('owid-covid-data.csv')
csvreader = csv.reader(file)
```

Retrieve all countries from Europe category

```python
rows = []
countries = []
numRow = 0

for row in csvreader:
    if numRow>0:
        rows.append(row)
    else:
        numRow = numRow + 1

for row in rows:
    if(row[1].strip().lower() == "europe"):
        if(row[2] not in countries):
            countries.append(row[2])

print(countries)
```

Go through the csv again to get data sort according to the year and month and respective details

```python
countryDict = {}

for country in countries:
    countryDict[country] = {}
    for row in rows:
        dateData = row[3]
        dateSplit = dateData.split("/")
        month = int(dateSplit[0])
        year = int(dateSplit[2])
        try:
            yearData = countryDict[country][year]
        except:
            templateData = {"Q1":{"stringencyTotal":0, "stringencyReadings":0, "covidCases":0, "death":0, "stringencyAvg":0},"Q2":{"stringencyTotal":0, "stringencyReadings":0, "covidCases":0, "death":0, "stringencyAvg":0},"Q3":{"stringencyTotal":0, "stringencyReadings":0, "covidCases":0, "death":0, "stringencyAvg":0},"Q4":{"stringencyTotal":0, "stringencyReadings":0, "covidCases":0, "death":0, "stringencyAvg":0}}
            countryDict[country][year] = templateData
        
        stringencyIndex = 0.00
        covidCaseNum = 0
        deaths = 0
        try:
            deaths = int(row[8])
        except:
            deaths = 0
        try:
            covidCaseNum = int(row[5])
        except:
            covidCaseNum = 0
        try:
            stringencyIndex = float(row[47])
        except:
            stringencyIndex = 0.00

        if month <= 3:
            currentTotal = countryDict[country][year]["Q1"]["stringencyTotal"]
            currentReading = countryDict[country][year]["Q1"]["stringencyReadings"]
            countryDict[country][year]["Q1"]["stringencyTotal"] = currentTotal + stringencyIndex
            countryDict[country][year]["Q1"]["stringencyReadings"] = currentReading + 1

            currentCovidCase = countryDict[country][year]["Q1"]["covidCases"]
            countryDict[country][year]["Q1"]["covidCases"] = currentCovidCase + covidCaseNum
            currentDeath = countryDict[country][year]["Q1"]["death"]
            countryDict[country][year]["Q1"]["death"] = currentDeath + deaths

        elif month>3 and month<=6:
            currentTotal = countryDict[country][year]["Q2"]["stringencyTotal"]
            currentReading = countryDict[country][year]["Q2"]["stringencyReadings"]
            countryDict[country][year]["Q2"]["stringencyTotal"] = currentTotal + stringencyIndex
            countryDict[country][year]["Q2"]["stringencyReadings"] = currentReading + 1

            currentCovidCase = countryDict[country][year]["Q2"]["covidCases"]
            countryDict[country][year]["Q2"]["covidCases"] = currentCovidCase + covidCaseNum
            currentDeath = countryDict[country][year]["Q2"]["death"]
            countryDict[country][year]["Q2"]["death"] = currentDeath + deaths
        
        elif month>6 and month<=9:
            currentTotal = countryDict[country][year]["Q3"]["stringencyTotal"]
            currentReading = countryDict[country][year]["Q3"]["stringencyReadings"]
            countryDict[country][year]["Q3"]["stringencyTotal"] = currentTotal + stringencyIndex
            countryDict[country][year]["Q3"]["stringencyReadings"] = currentReading + 1

            currentCovidCase = countryDict[country][year]["Q3"]["covidCases"]
            countryDict[country][year]["Q3"]["covidCases"] = currentCovidCase + covidCaseNum
            currentDeath = countryDict[country][year]["Q3"]["death"]
            countryDict[country][year]["Q3"]["death"] = currentDeath + deaths
        
        else:
            currentTotal = countryDict[country][year]["Q4"]["stringencyTotal"]
            currentReading = countryDict[country][year]["Q4"]["stringencyReadings"]
            countryDict[country][year]["Q4"]["stringencyTotal"] = currentTotal + stringencyIndex
            countryDict[country][year]["Q4"]["stringencyReadings"] = currentReading + 1

            currentCovidCase = countryDict[country][year]["Q4"]["covidCases"]
            countryDict[country][year]["Q4"]["covidCases"] = currentCovidCase + covidCaseNum
            currentDeath = countryDict[country][year]["Q4"]["death"]
            countryDict[country][year]["Q4"]["death"] = currentDeath + deaths
print(countryDict)
```

Calculate average stringency

```python
allCountries = countryDict.keys()
for country in allCountries:
    yearColl = countryDict[country].keys()
    for year in yearColl:
        quarterColl = countryDict[country][year].keys()
        for quarter in quarterColl:
            try:
                countryDict[country][year][quarter]["stringencyAvg"] = round(countryDict[country][year][quarter]["stringencyTotal"]/countryDict[country][year][quarter]["stringencyReadings"], 2)
            except:
                countryDict[country][year][quarter]["stringencyAvg"] = 0
print(countryDict)
```

# Prepare API

API will receive a GET request and return the dictionary as a JSON

```python
app = flask.Flask(__name__)

@app.route('/', methods=['GET'])
def displayData():
    response = flask.jsonify(countryDict)
    response.headers.add('Access-Control-Allow-Origin', '*')
    return response

app.run()
```
